```python
import numpy as np

# Para reprodutibilidade
SEED = 42
rng = np.random.default_rng(SEED)

print("--- Soluções dos Exercícios NumPy ---")
print("\n## Fáceis (24)")

# ----------------------------------------------------
# 1. Vetor de zeros
# ----------------------------------------------------
print("\n1. Vetor de zeros")
Z = np.zeros(10)
print(f"Array: {Z}")
print(f"dtype: {Z.dtype}")
print(f"shape: {Z.shape}")

# ----------------------------------------------------
# 2. Vetor de uns
# ----------------------------------------------------
print("\n2. Vetor de uns")
O = np.ones(10, dtype=np.float64)
# Alternativa: O = np.ones(10).astype(np.float64)
print(f"Array: {O}")
print(f"dtype: {O.dtype}")

# ----------------------------------------------------
# 3. Sequência 0–9
# ----------------------------------------------------
print("\n3. Sequência 0–9")
S = np.arange(10)
print(f"Array: {S}")

# ----------------------------------------------------
# 4. Matriz 3×3 com 0…8
# ----------------------------------------------------
print("\n4. Matriz 3×3 com 0…8")
M = np.arange(9).reshape(3, 3)
print("Matriz 3x3:")
print(M)

# ----------------------------------------------------
# 5. Índices não zero
# ----------------------------------------------------
print("\n5. Índices não zero")
v = np.array([1, 2, 0, 0, 4, 0])
idx = np.nonzero(v)
# Alternativa: idx = np.where(v != 0)
print(f"Vetor: {v}")
print(f"Índices não nulos: {idx}")

# ----------------------------------------------------
# 6. Linspace
# ----------------------------------------------------
print("\n6. Linspace")
L = np.linspace(0, 1, 12)
print(f"Linspace: {L}")

# ----------------------------------------------------
# 7. Identidade 5×5
# ----------------------------------------------------
print("\n7. Identidade 5×5")
I = np.eye(5)
print("Matriz Identidade 5x5:")
print(I)

# ----------------------------------------------------
# 8. Aleatórios 3×3×3
# ----------------------------------------------------
print("\n8. Aleatórios 3×3×3")
R = rng.random((3, 3, 3))
print("Array 3x3x3 de aleatórios:")
print(R)

# ----------------------------------------------------
# 9. Min/Max
# ----------------------------------------------------
print("\n9. Min/Max")
R2 = rng.random((10, 10))
min_val = R2.min()
max_val = R2.max()
print(f"Mínimo: {min_val:.4f}")
print(f"Máximo: {max_val:.4f}")

# ----------------------------------------------------
# 10. Padronização 1D
# ----------------------------------------------------
print("\n10. Padronização 1D")
x = np.array([1, 2, 3, 4, 5], dtype=float)
x_std = (x - x.mean()) / x.std()
print(f"Vetor original: {x}")
print(f"Vetor padronizado: {x_std}")
print(f"Média padronizada (aprox. 0): {x_std.mean():.2e}")
print(f"Desvio padrão padronizado (aprox. 1): {x_std.std():.2e}")

# ----------------------------------------------------
# 11. Conversão de tipos
# ----------------------------------------------------
print("\n11. Conversão de tipos")
arr_int32 = np.array([1, 2, 3], dtype=np.int32)
arr_float32 = arr_int32.astype(np.float32)
arr_int64 = arr_float32.astype(np.int64)
print(f"Int32: {arr_int32}, dtype: {arr_int32.dtype}")
print(f"Float32: {arr_float32}, dtype: {arr_float32.dtype}")
print(f"Int64: {arr_int64}, dtype: {arr_int64.dtype}")

# ----------------------------------------------------
# 12. Clipping
# ----------------------------------------------------
print("\n12. Clipping")
a = np.array([2, 5, 7, 1, 9])
a_clipped = np.clip(a, 0, 5)
print(f"Original: {a}")
print(f"Clipping [0, 5]: {a_clipped}")

# ----------------------------------------------------
# 13. Inversão
# ----------------------------------------------------
print("\n13. Inversão")
v_orig = np.arange(10)
v_inv = np.flip(v_orig, axis=0)
print(f"Original: {v_orig}")
print(f"Invertido: {v_inv}")

# ----------------------------------------------------
# 14. Empilhamento
# ----------------------------------------------------
print("\n14. Empilhamento")
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
H = np.hstack((a, b))  # Alternativa: np.concatenate((a, b), axis=0)
V = np.vstack((a, b))  # Alternativa: np.concatenate((a[None,:], b[None,:]), axis=0)
print(f"Original a: {a}, b: {b}")
print(f"Horizontal (hstack): {H}")
print("Vertical (vstack):\n", V)

# ----------------------------------------------------
# 15. Diagonal
# ----------------------------------------------------
print("\n15. Diagonal")
D = np.diag([5] * 4)  # Cria um array 1D de 4x 5, e depois usa diag
# Alternativa para a matriz 4x4 de zeros, preenchendo a diagonal:
# D = np.zeros((4,4))
# np.fill_diagonal(D, 5)
print("Matriz com diagonal 5s:")
print(D)

# ----------------------------------------------------
# 16. Somas por eixo
# ----------------------------------------------------
print("\n16. Somas por eixo")
A = np.arange(12).reshape(3, 4)
sum_rows = A.sum(axis=1)
sum_cols = A.sum(axis=0)
print("Matriz A:")
print(A)
print(f"Soma por linhas (axis=1): {sum_rows}")
print(f"Soma por colunas (axis=0): {sum_cols}")

# ----------------------------------------------------
# 17. Frequências inteiras
# ----------------------------------------------------
print("\n17. Frequências inteiras")
R_int = rng.integers(0, 10, size=100)
# np.bincount é mais eficiente para inteiros não negativos (0 a n)
counts = np.bincount(R_int, minlength=10)
print(f"Array de 100 inteiros (primeiros 10): {R_int[:10]}...")
print(f"Frequências (0 a 9): {counts}")

# ----------------------------------------------------
# 18. Slicing com passo
# ----------------------------------------------------
print("\n18. Slicing com passo")
v = np.arange(20)
v_slice = v[1::3]
print(f"Vetor: {v}")
print(f"Slicing [1::3]: {v_slice}")

# ----------------------------------------------------
# 19. Máscara booleana
# ----------------------------------------------------
print("\n19. Máscara booleana")
v_all = np.arange(20)
mask_par = (v_all % 2 == 0)
v_par = v_all[mask_par]
print(f"Máscara para pares: {mask_par[:10]}...")
print(f"Pares (filtrados): {v_par}")

# ----------------------------------------------------
# 20. Lista de listas → array
# ----------------------------------------------------
print("\n20. Lista de listas → array")
list_of_lists = [[1, 2, 3], [4, 5, 6]]
arr_from_list = np.array(list_of_lists)
print("Array convertido:")
print(arr_from_list)
print(f"shape: {arr_from_list.shape}")
print(f"dtype: {arr_from_list.dtype}")

# ----------------------------------------------------
# 21. Datas básicas
# ----------------------------------------------------
print("\n21. Datas básicas")
start_date = np.datetime64('2025-01-01')
dates = np.arange(start_date, start_date + np.timedelta64(7, 'D'))
# Alternativa: np.arange('2025-01-01', '2025-01-08', dtype='datetime64[D]')
print(f"Array de datas: {dates}")

# ----------------------------------------------------
# 22. Centralizar colunas
# ----------------------------------------------------
print("\n22. Centralizar colunas")
A_orig = np.array([[1., 2., 3.], [4., 5., 6.]])
A_mean = A_orig.mean(axis=0)  # Média de cada coluna (vetor 1D de 3 elementos)
A_cent = A_orig - A_mean  # Broadcasting
print("Matriz A original:")
print(A_orig)
print(f"Médias por coluna: {A_mean}")
print("Matriz A centralizada:")
print(A_cent)
print(f"Novas médias (aprox. 0): {A_cent.mean(axis=0)}")

# ----------------------------------------------------
# 23. Produto elemento-a-elemento vs. escalar
# ----------------------------------------------------
print("\n23. Produto elemento-a-elemento vs. escalar")
a_vec = np.array([1, 2, 3])
b_vec = np.array([4, 5, 6])
prod_elem = a_vec * b_vec  # Produto elemento a elemento
prod_dot = np.dot(a_vec, b_vec)  # Produto escalar (dot product)
# Alternativa: prod_dot = a_vec @ b_vec
print(f"a: {a_vec}, b: {b_vec}")
print(f"Produto elemento-a-elemento: {prod_elem}")
print(f"Produto escalar (dot product): {prod_dot}")

# ----------------------------------------------------
# 24. Troca de colunas
# ----------------------------------------------------
print("\n24. Troca de colunas")
A_swap = np.arange(12).reshape(3, 4)
print("Matriz A original:")
print(A_swap)
# Troca as colunas 0 e 2 (cópia mínima)
A_swapped = A_swap[:, [2, 1, 0, 3]]
print("Matriz A com colunas 0 e 2 trocadas:")
print(A_swapped)

print("\n## Médios (12)")

# ----------------------------------------------------
# 25. Flatten vs. ravel
# ----------------------------------------------------
print("\n25. Flatten vs. ravel")
M_orig = np.arange(4).reshape(2, 2)
M_flat = M_orig.flatten()  # Cria CÓPIA
M_ravel = M_orig.ravel()  # Cria VIEW (se possível)

# Modificando M_ravel (a view):
M_ravel[0] = 99
print("Matriz original (M_orig):")
print(M_orig)  # ORIGINAL MUDOU, pois M_ravel é uma VIEW
print(f"M_ravel (VIEW): {M_ravel}")
print(f"M_flat (CÓPIA): {M_flat}")  # CÓPIA não mudou

# ----------------------------------------------------
# 26. Tabuleiro 8×8
# ----------------------------------------------------
print("\n26. Tabuleiro 8×8")
C = np.zeros((8, 8), dtype=int)
C[1::2, ::2] = 1  # Linhas ímpares, colunas pares
C[::2, 1::2] = 1  # Linhas pares, colunas ímpares
print("Tabuleiro 8x8:")
print(C)

# ----------------------------------------------------
# 27. np.where condicional
# ----------------------------------------------------
print("\n27. np.where condicional")
x = np.arange(-5, 6)
x_mod = np.where(x < 0, 0, x)
print(f"Original: {x}")
print(f"Modificado: {x_mod}")

# ----------------------------------------------------
# 28. Precisão de média
# ----------------------------------------------------
print("\n28. Precisão de média")
N = 1_000_000
R_uni = rng.uniform(0, 1, size=N)
mean_calc = R_uni.mean()
mean_theory = 0.5
error = np.abs(mean_calc - mean_theory)
print(f"Média calculada: {mean_calc:.6f}")
print(f"Média teórica: {mean_theory:.6f}")
print(f"Erro absoluto: {error:.2e}")

# ----------------------------------------------------
# 29. Produto cartesiano (broadcasting)
# ----------------------------------------------------
print("\n29. Produto cartesiano (broadcasting)")
a_broad = np.array([1, 2, 3])
b_broad = np.array([10, 20, 30, 40])
# a_broad é (3,) -> (3, 1) para broadcasting
# b_broad é (4,) -> (1, 4) para broadcasting
A_i = a_broad[:, np.newaxis]
B_j = b_broad[np.newaxis, :]
# Alternativa: A_i = a_broad.reshape(-1, 1)
Somas = A_i + B_j
print(f"A_i.shape: {A_i.shape}, B_j.shape: {B_j.shape}")
print("Matriz de somas (3x4):")
print(Somas)

# ----------------------------------------------------
# 30. Convolução 1D simples
# ----------------------------------------------------
print("\n30. Convolução 1D simples")
# Requer np.lib.stride_tricks
from numpy.lib.stride_tricks import sliding_window_view

s = np.array([1, 2, 3, 4, 5])
k = np.array([0.5, 1.0, 0.5])  # Kernel de tamanho 3
k_flip = np.flip(k)

# 1. Crie as janelas deslizantes (rolling window) de s, de tamanho len(k)
windows = sliding_window_view(s, window_shape=len(k))

# 2. Multiplicação matricial (dot product de cada janela pelo kernel invertido)
# 'windows' é (N-M+1) x M. 'k_flip' é M x 1 (implicitamente)
# windows @ k_flip
conv_result = windows @ k_flip

# O resultado deve ser o mesmo que np.convolve(s, k, mode='valid')
print(f"Sinal (s): {s}")
print(f"Kernel (k): {k}")
print(f"Resultado da Convolução 'valid': {conv_result}")
# print(f"Checagem com np.convolve: {np.convolve(s, k, mode='valid')}")

# ----------------------------------------------------
# 31. Normalização min–max por coluna
# ----------------------------------------------------
print("\n31. Normalização min–max por coluna")
A = rng.normal(size=(5, 4))
# Simule uma coluna constante
A[:, 0] = 5.0
print("Matriz A (primeiras 2 linhas):")
print(A[:2])

A_min = A.min(axis=0)
A_max = A.max(axis=0)
A_range = A_max - A_min

# Máscara para colunas constantes (range == 0)
is_constant = A_range == 0

# Normalização: (X - min) / range
# Use np.divide com 'where' para evitar divisão por zero
A_norm = np.divide(A - A_min, A_range, out=np.zeros_like(A), where=~is_constant)

# Colunas constantes (range == 0): resultado é 0.0 (pela cláusula out=np.zeros_like(A))
# Alternativamente, para colunas constantes, o resultado deveria ser 0.5 ou 0, dependendo da convenção.
# Com o 'out' zerado, a coluna constante A[:, 0] resultará em 0s.

print("Matriz A Normalizada [0,1] (primeiras 2 linhas):")
print(A_norm[:2])
print(f"Mínimos (por coluna): {A_norm.min(axis=0)}")
print(f"Máximos (por coluna): {A_norm.max(axis=0)}")

# ----------------------------------------------------
# 32. Ordenar por coluna
# ----------------------------------------------------
print("\n32. Ordenar por coluna")
A_sort = np.array([[3, 9], [1, 4], [2, 5]])
print("Matriz A original:")
print(A_sort)

# Pega os índices que ordenam a 2ª coluna (índice 1)
sort_idx = np.argsort(A_sort[:, 1])

# Usa os índices para reordenar as linhas de A
A_sorted = A_sort[sort_idx]
print("Matriz A ordenada pela 2ª coluna:")
print(A_sorted)

# ----------------------------------------------------
# 33. Únicos + contagens + primeira ocorrência
# ----------------------------------------------------
print("\n33. Únicos + contagens + primeira ocorrência")
x_unique = np.array([3, 1, 2, 3, 2, 3, 1])

# Retorna únicos, índices da primeira ocorrência, e contagens
unique_vals, first_indices, counts = np.unique(
    x_unique, return_index=True, return_counts=True
)

print(f"Vetor: {x_unique}")
print(f"Valores únicos: {unique_vals}")
print(f"Contagens: {counts}")
print(f"Índices da 1ª ocorrência: {first_indices}")

# ----------------------------------------------------
# 34. Matriz de distâncias 2D
# ----------------------------------------------------
print("\n34. Matriz de distâncias 2D")
# N=3 pontos em 2D
P = np.array([[1, 1], [4, 5], [0, 5]])
print(f"Pontos P (N={P.shape[0]}):")
print(P)

# Broadcasting: (N, 1, 2) - (1, N, 2) -> (N, N, 2) (diferenças)
diffs = P[:, np.newaxis, :] - P[np.newaxis, :, :]

# Quadrado das diferenças
sq_diffs = diffs**2  # (N, N, 2)

# Soma ao longo do eixo das coordenadas (axis=2)
sum_sq_diffs = sq_diffs.sum(axis=2)  # (N, N) (Distância Euclidiana ao quadrado)

# Raiz quadrada
dist_matrix = np.sqrt(sum_sq_diffs)

print("Matriz de Distâncias Euclidianas:")
print(dist_matrix)

# ----------------------------------------------------
# 35. Covariância e correlação
# ----------------------------------------------------
print("\n35. Covariância e correlação")
# Matriz 1000x3 de N(0,1)
X_norm = rng.standard_normal(size=(1000, 3))

# `np.cov` assume que as *linhas* são as variáveis por padrão (rowvar=True).
# Como as *colunas* são as variáveis, usamos `rowvar=False`.
cov_matrix = np.cov(X_norm, rowvar=False)

# `np.corrcoef` assume que as *linhas* são as variáveis por padrão.
# Para colunas como variáveis, transpomos antes (X_norm.T) ou usamos a matriz de covariância.
corr_matrix = np.corrcoef(X_norm, rowvar=False)

print("Matriz de Covariância (3x3):")
print(cov_matrix)
print("\nMatriz de Correlação (3x3):")
print(corr_matrix)

# ----------------------------------------------------
# 36. One-hot encoding
# ----------------------------------------------------
print("\n36. One-hot encoding")
y = np.array([2, 0, 1, 2])
n_classes = y.max() + 1
n_samples = len(y)

# Cria um array de índices de coluna
col_indices = y
# Cria um array de índices de linha (0 a n_samples-1)
row_indices = np.arange(n_samples)

# Inicializa matriz de zeros (N x K)
one_hot = np.zeros((n_samples, n_classes), dtype=int)

# Usa fancy indexing para definir 1 nas posições corretas
one_hot[row_indices, col_indices] = 1

print(f"Rótulos y: {y}")
print("One-hot Encoding (4x3):")
print(one_hot)

print("\n## Difíceis (6)")

# ----------------------------------------------------
# 37. k-means 2D (NumPy puro)
# ----------------------------------------------------
print("\n37. k-means 2D (NumPy puro)")
K = 3
N_POINTS = 100
MAX_ITERS = 10

# Gere dados (3 clusters simulados)
X_data = np.vstack([
    rng.normal(loc=[0, 0], scale=1, size=(N_POINTS // 3, 2)),
    rng.normal(loc=[5, 5], scale=1, size=(N_POINTS // 3, 2)),
    rng.normal(loc=[-5, 5], scale=1, size=(N_POINTS - 2 * (N_POINTS // 3), 2))
])

# 1. Inicialize centróides K centróides aleatoriamente (selecione K pontos de X_data)
init_idx = rng.choice(N_POINTS, size=K, replace=False)
centroids = X_data[init_idx]
print(f"Centróides iniciais:\n{centroids}")

for i in range(MAX_ITERS):
    # --- Passo 1: Atribuição ---
    # Broadcasting: (N, 1, 2) - (1, K, 2) -> (N, K, 2) (diferenças)
    # diffs = X_data[:, np.newaxis, :] - centroids[np.newaxis, :, :]

    # Distância ao quadrado (otimizado)
    # (N, K) matriz de distâncias ao quadrado entre cada ponto e cada centróide
    # numpy.einsum('...ij, ...ij -> ...i', a, a)
    distances_sq = np.sum((X_data[:, np.newaxis, :] - centroids)**2, axis=2)

    # Encontre o centróide mais próximo
    # assignments é um vetor de N elementos (0, 1, ou 2)
    assignments = np.argmin(distances_sq, axis=1)

    # --- Passo 2: Atualização de Centróides ---
    new_centroids = np.zeros((K, 2))

    for k in range(K):
        # Selecione os pontos atribuídos ao cluster k
        points_in_cluster = X_data[assignments == k]

        if len(points_in_cluster) > 0:
            # Calcule a nova média
            new_centroids[k] = points_in_cluster.mean(axis=0)
        else:
            # Se um cluster estiver vazio, mantenha o centróide ou re-inicialize
            new_centroids[k] = centroids[k]

    # Critério de parada (se os centróides não mudarem)
    if np.allclose(centroids, new_centroids):
        print(f"Convergiu na iteração {i + 1}.")
        break

    centroids = new_centroids

print(f"Centróides finais (após {i + 1} iterações):\n{centroids}")
# print(f"Atribuições (primeiras 10): {assignments[:10]}")

# ----------------------------------------------------
# 38. PCA básica (autovetores)
# ----------------------------------------------------
print("\n38. PCA básica (autovetores)")
# Dados 2D (simulados com alguma correlação)
N_PCA = 100
X_orig = np.dot(rng.standard_normal(size=(N_PCA, 2)), [[2, 1], [0, 0.5]])
print(f"X.shape: {X_orig.shape}")

# 1. Centralizar X
X_centered = X_orig - X_orig.mean(axis=0)

# 2. Matriz de Covariância (rowvar=False)
C_matrix = np.cov(X_centered, rowvar=False)

# 3. Autovalores (lambdas) e Autovetores (V)
# V é tal que C @ V[:,i] = lambdas[i] * V[:,i]
lambdas, V = np.linalg.eigh(C_matrix)

# 4. Selecionar o Autovetor Principal (o que tem o maior autovalor)
# np.eigh garante que autovalores vêm em ordem crescente
principal_comp_idx = np.argmax(lambdas)
W = V[:, principal_comp_idx] # Vetor de projeção (1ª componente principal)

# 5. Projeção 2D -> 1D
# X_centered @ W
X_proj = X_centered @ W[:, np.newaxis] # (N, 2) @ (2, 1) -> (N, 1)

# 6. Reconstrução aproximada 1D -> 2D
# X_proj @ W.T + média original
X_reconst = X_proj @ W[np.newaxis, :] + X_orig.mean(axis=0)

print(f"Matriz de Covariância:\n{C_matrix}")
print(f"Autovalores: {lambdas}")
print(f"Vetor principal (W): {W}")
print(f"X_proj.shape (1D): {X_proj.shape}")
print(f"X_reconst.shape: {X_reconst.shape}")

# ----------------------------------------------------
# 39. Convolução 2D (im2col)
# ----------------------------------------------------
print("\n39. Convolução 2D (im2col)")
# Requer np.lib.stride_tricks
from numpy.lib.stride_tricks import sliding_window_view

# Imagem 4x4
I = np.arange(16).reshape(4, 4)
# Kernel 3x3
K = np.ones((3, 3)) * 0.1
H_K, W_K = K.shape

print("Imagem I (4x4):")
print(I)

# 1. Gerar blocos (im2col)
# O sliding_window_view retorna (H_out, W_out, H_k, W_k)
blocks = sliding_window_view(I, window_shape=(H_K, W_K))
H_out, W_out, _, _ = blocks.shape

# 2. Reshape dos blocos para multiplicação matricial (im2col)
# (H_out * W_out) x (H_k * W_k)
blocks_flat = blocks.reshape(H_out * W_out, H_K * W_K)

# 3. Reshape do kernel (col2im)
# (H_k * W_k) x 1
K_flat = K.flatten()[:, np.newaxis]

# 4. Multiplicação matricial (H_out * W_out) x 1
Conv_flat = blocks_flat @ K_flat

# 5. Reshape do resultado para o formato de saída (H_out, W_out)
Conv_result = Conv_flat.reshape(H_out, W_out)

print(f"Blocks.shape: {blocks.shape}")
print(f"Blocks_flat.shape (im2col): {blocks_flat.shape}")
print("Resultado da Convolução 'valid' (2x2):")
print(Conv_result)

# ----------------------------------------------------
# 40. Sistemas lineares e resíduo
# ----------------------------------------------------
print("\n40. Sistemas lineares e resíduo")
N_sys = 100
# Gere A bem-condicionada (por exemplo, matriz aleatória com diagonal dominante)
A = rng.random(size=(N_sys, N_sys))
A = A + N_sys * np.eye(N_sys) # Garante que A é bem-condicionada
b = rng.random(size=N_sys)

# 1. Resolva Ax=b
x = np.linalg.solve(A, b)

# 2. Calcule resíduo r = Ax - b
r = A @ x - b

# 3. Calcule o resíduo relativo ||Ax−b||/||b||
residuo_relativo = np.linalg.norm(r) / np.linalg.norm(b)

# 4. Calcule cond(A)
cond_A = np.linalg.cond(A)

print(f"A.shape: {A.shape}")
print(f"Resíduo relativo ||Ax−b||/||b||: {residuo_relativo:.2e}")
print(f"Número de condição cond(A): {cond_A:.2e}")

# ----------------------------------------------------
# 41. Regressão linear (OLS) com NumPy
# ----------------------------------------------------
print("\n41. Regressão linear (OLS) com NumPy")
N_OLS = 100
# Dados de exemplo: y = 2x + 1 + ruído
X_feat = np.linspace(0, 10, N_OLS)[:, np.newaxis] # (N, 1)
noise = rng.normal(0, 0.5, size=(N_OLS, 1))
y_target = 2 * X_feat + 1 + noise

# Adiciona coluna de 1s para o intercepto (X_design)
X_design = np.hstack([X_feat, np.ones((N_OLS, 1))])

# 1. Calcule coeficientes OLS usando np.linalg.lstsq
# lstsq retorna: solução, soma dos resíduos quadrados, posto, valores singulares
# A solução `beta` contém [coeficiente, intercepto]
beta, residuals, rank, s = np.linalg.lstsq(X_design, y_target, rcond=None)
beta = beta.flatten() # [coef_x, intercepto]

# 2. Avalie y_pred (treino)
y_pred = X_design @ beta[:, np.newaxis]

# 3. Calcule RMSE
mse = np.mean((y_pred - y_target)**2)
rmse = np.sqrt(mse)

print(f"y = {beta[0]:.2f} * x + {beta[1]:.2f}")
print(f"RMSE (treino): {rmse:.4f}")

# ----------------------------------------------------
# 42. Rolling window 2D com strides
# ----------------------------------------------------
print("\n42. Rolling window 2D com strides")
# Requer np.lib.stride_tricks
from numpy.lib.stride_tricks import sliding_window_view

# Imagem 6x6
IMG = np.arange(36).reshape(6, 6)
WINDOW_SIZE = (3, 3)

print("Imagem original (6x6):")
print(IMG)

# 1. Gerar a janela deslizante 3x3
# O resultado terá shape (H_out, W_out, 3, 3)
windows_3x3 = sliding_window_view(IMG, window_shape=WINDOW_SIZE)

# 2. Calcule a média local ao longo dos eixos da janela (eixos 2 e 3)
# O resultado terá shape (H_out, W_out)
local_mean = windows_3x3.mean(axis=(2, 3))

print(f"Shape da janela deslizante: {windows_3x3.shape}")
print("Média local 3x3 (4x4):")
print(local_mean)
```
